```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

This model is created to predict the salary of an individual given his years of experience and his scores in test and interview. This is a multivariate linear regression model as the dependent variable (salary) compared to the independent variable is positively correlative.
There are also some data preprocessing carried out on the data before the model is built.

```python
df = pd.read_csv('hiring.csv')
df
```

The data needs some cleaning. The experience column should be in years which should be an integer. There are missing values which might indicate entry level (no exp). for testscore. There is a missing value. This could be replaced with the mean.

```python
df.experience = df['experience'].replace(['five','two','seven','three','ten','eleven'],[5,2,7,3,10,11])
df
```

replaced the word text with numbers. I could have used a module called word2numbers for this activity is the data was large. But I am good to go with the replace function.

```python
df.experience = df.experience.fillna(0)
df
```

```python
import math
av = math.floor(df['test_score(out of 10)'].mean())
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(av)
print(df)
```

The data has been cleaned completely and can now be used in my analysis. The below cell creates my linear regression object and fits my data in from my dataframe

```python
reg = linear_model.LinearRegression()
reg.fit(df[['experience','test_score(out of 10)','interview_score(out of 10)']], df['salary($)'])
```

I have to predict salary for two individuals.
1. 2 years experience, 9 test score, 6 interview
2. 12 years experience, 10 test score, 10 interview



```python
import math
guy1 = reg.predict([[2,9,6]])
print(f'The man with 2 years experience and scored 9 and 6 on the test and interview respectively salary would be {math.floor(guy1)}')
guy2 = reg.predict([[12,10,10]])
print(f'The man with 12 years experience and scored 10 and 10 on the test and interview respectively salary would be {math.floor(guy2)}')
```
